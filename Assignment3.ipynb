```python
import random
from collections import Counter

def hat():
    # Create a list of natural numbers up to 2024 
    naturals = list(range(1, 2025))

    # Continue until one number is left
    while len(naturals) > 1:
        # Shuffle the list and select the first two numbers from the list 
        random.shuffle(naturals)
        nat1 = naturals[0]
        nat2 = naturals[1]
        difference = abs(nat1 - nat2)

        # Add the difference to the list and remove the two chosen numbers
        naturals.append(difference)
        naturals.remove(nat1)
        naturals.remove(nat2)
        
    #When length=1 take the only number in the list    
    final_number = naturals[0]
    return final_number

# Repeat function
def repeatHat():
    # Repeats the game a certain number of times
    results = []
    for _ in range(100):
        result = hat()
        results.append(result)
    
    # Count the frequency of each result
    frequencies = Counter(results)
    for number, count in sorted(frequencies.items()):
        print(f"{number}- {count}")

if __name__ == "__main__":
    repeatHat()
```
